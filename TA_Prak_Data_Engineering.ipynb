```python
import io
df = pd.read_csv(io.StringIO(uploaded['Survey Pekerjaan Impian.csv'].decode('utf-8')))
```

```python
df.isnull().sum()
```

Membuat 1 kolom baru bernama "Kota Destructured" yg berisi destructuring data kota impian yang memiliki beberapa value sekaligus menjadi satu value saja

```python
new_data = df['Kota Impian Tempat Anda Bekerja'].str.split(';', expand=True).stack().reset_index(level=1, drop=True).to_frame('Kota Destructured')
new_data
```

Menggabungkan kolom baru "Kota Destructured" ke dataset

```python
df = pd.concat([df, new_data], axis=1)
df
```
